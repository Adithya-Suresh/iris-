```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(12,12)
data = pd.read_csv('https://raw.githubusercontent.com/Adithya-Suresh/iris-/master/Iris.csv')
print(data.head(10))
```

```
print(data.info())
```

```
print(data.describe())
```

```
data.plot.scatter(x='SepalLengthCm',y='SepalWidthCm')
```

```
data.plot.scatter(x='PetalLengthCm',y='PetalWidthCm')
```

```
dte=data
dte=dte.drop(dte.index[dte.SepalLengthCm>=7.0],axis=0)

dte=dte.drop(dte.index[dte.SepalWidthCm>=3.5],axis=0)
dte.plot.scatter(x='SepalLengthCm',y='SepalWidthCm')
dte.plot.scatter(x='PetalLengthCm',y='PetalWidthCm')
print(dte.describe())
```

```
#h={'Species':{'Iris-setosa':'0','Iris-versicolor':'1','Iris-virginica':'2'}}
d=pd.DataFrame(dte)
#d.replace(h,inplace=True)
d=d.drop(['Id'],axis=1)
print(d.head(100))
```

```
X=d
X=X.drop(['Species'],axis=1)

#from sklearn.preprocessing import StandardScaler
#sc = StandardScaler()
#X = sc.fit_transform(X)
y=d['Species'].values
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.50, random_state = 0)
```

```
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 40, criterion = 'entropy', random_state = 0)
classifier.fit(X_train,y_train)
y_p=classifier.predict(X_test)
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_p)
print(acc)
```

```
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_p)
labels =['Pr0', 'Pr1','Pr2']
print(*labels)
for line in cm:
    print(*line)
```
